```python
class ActionInstance():
    count = 0
    
    def __init__(self, action_label, video_id, start, end, duration=0):
        self.action_label = action_label
        self.video_id = video_id
        self.start = start
        self.end = end
        self.duration = self.end - self.start
        self.coverage = self.duration / duration if duration != 0 else None
    @classmethod
    def inc_count(cls):
        cls.count += 1
    @classmethod
    def reset_count(cls):
        cls.count = 0

class EventInstance(ActionInstance):
    def __init__(self, action_label, video_id, start, end, duration=0):
        super().__init__(action_label, video_id, start, end, duration)
        self.actions = []
        
    def add_action(self, action_instance):
        self.actions.append(action_instance)
    
    def get_num_seg(self):
        return len(self.actions)

    
```

```python
action = ActionInstance(1, 'dsad', 1, 10, 20)
action.coverage

event = EventInstance(2, 'fuck', 1, 21, 50)
event.add_action(action)
```

```python
class VideoInstance():
    count = 0
    
    def __init__(self,
                 video_id,
                 duration=None,
                 subset=None,
                 resolution=None,
                 url=None,
                 **kwargs):
        self.events = []
        self.video_id = video_id
        self.duration = duration
        self.subset = subset
        self.resolution = resolution
        self.url = url
        
    @classmethod
    def inc_count(cls):
        cls.count += 1
    @classmethod
    def reset_count(cls):
        cls.count = 0
        
    def add_event(self, event_instance):
        self.events.append(event_instance)
    
    def get_event_num(self):
        return len(self.events)
```

```python
# Now read all the action instance and get some insights of them
all_actions = []
all_events = []
all_videos = []
ActionInstance.reset_count()
EventInstance.reset_count()
VideoInstance.reset_count()

for video_id, video_info in annotation.items():
    new_video = VideoInstance(video_id)
    VideoInstance.inc_count()
    for event_id, event_info in video_info.items():
        new_event = EventInstance(event_info['event'],
                                  video_id,
                                  float(event_info['timestamps'][0][0]),
                                  float(event_info['timestamps'][0][1]),
                                  )
        if event_info['segments'] is not None:
            for action_id, action_info in event_info['segments'].items():
                new_action = ActionInstance(action_info['stages'],
                                           event_id,
                                           float(action_info['timestamps'][0][0]),
                                           float(action_info['timestamps'][0][1]),
                                           new_event.duration)
                new_event.add_action(new_action)
                all_actions.append(new_action)
                ActionInstance.inc_count()

        new_video.add_event(new_event)
        EventInstance.inc_count()
        all_events.append(new_event)
        
    all_videos.append(new_video)
        
assert len(all_actions) == ActionInstance.count
assert len(all_events) == EventInstance.count
assert len(all_videos) == VideoInstance.count

print(f'There are {ActionInstance.count} SubActions {EventInstance.count} Events and {VideoInstance.count} Videos.')
```
